Diccionario de diccionarios

```python
europe={
    "spain":{"population":10,"capital":"madrid"},
    "germany":{"population":10,"capital":"berlin"},
    "france":{"population":10,"capital":"paris"},
    "normay":{"population":10,"capital":"oslo"}}
print(europe)        
```

```python
countries=["spain","germany","france","norway"]
capitals=["madrid","berlin","paris","oslo"]
population=[90,42,64,68]
europe={'countries':countries, 'capitals':capitals, 'population':population}
print(europe)
```

```python
import pandas as pd
```

```python
df_europe = pd.DataFrame(europe)
print(df_europe)
```

```python
print(df_europe.index)
```

```python
row_labels=["SP","GE","FR","NO"]
df_europe.index=row_labels
print(df_europe.index)
```
