```python
import pandas as pd
```

```python
data = pd.read_csv('data_fanpage.csv')
```

```python
for x in data:
    print(x)
```

```python
for lab, row in data.iterrows():
    print(lab)
    print(row)
```

```python
for lab, row in data.iterrows():
    print(str(lab) + ":" + str(row["comments"]))
```

```python
for lab, row in data.iterrows():
    data.loc[lab, "comments * 2"] = row["comments"] * 2
```

```python
for x in data:
    print(x)
```

```python
for lab, row in data.iterrows():
    print(str(lab) + ":" + str(row["comments"]) + ":" + str(row["comments * 2"]))
```

```python
data["comments * 3"] = data["comments"].apply(lambda x: x * 3)
```

```python
for x in data:
    print(x)
```
